## 1-1. 시계열에 대한 소개 

시계열은 시간의 흐름에 따라 기록된 데이터를 의미합니다. 
- 따라서 현재와 과거의 데이터 사이에는 상관관계가 존재하게 됩니다.

<img src='./IMG/img1.png'>

예시 : 페이스북의 방문 시계열 데이터

시계열 예측이란 과거의 데이터를 기반으로 현재를 설명하고 미래를 예측하는 것을 의미합니다.

<img src = 'IMG/img2.png'>

예시 : 원유 수요 예측

거시 경제에 대한 지표들은 시계열 분석이 많이 적용되는 분야 (Econometrics)

- 추세(Trend)/ 계절성(Seasonality)/ 경기순환 (Cycle) 

<img src = 'IMG/img3.png'>

시계열이 회귀 분석(데이터)과 다른 점 

- x,y 사이의 관계 (correlation)와 달리 데이터(y)의 현재와 과거 사이의 관계 (autocorrelation) 

<img src = 'IMG/img4.png'>

데이터 분석은 수학적인 모형을 통해 관계를 설명 
- 다양한 형태의 모형이 가능 

1. 자기 상관 관계를 모형화 : ARIMA 
2. 시계열을 Trend & Seasonality로 분해 : 분해법 (Decomposition) 
3. 시계열 회귀 분석 : 회귀 모형과 시계열 모형을 함께 사용 
4. 분해법과 ARIMA를 함께 사용하는 모형 (TBATS) 
5. 신경망을 이용한 시계열 모형

## 1-4. 시계열의 구성요소 

### 추세 (Trend)와 계절성 (Seasonality) 

- `추세` : 데이터가 장기적으로 증가하거나 감소할 때, 추세(trend)가 존재한다고 말합니다. 추세는 반드시 선형적일 필요는 없습니다. 때때로 어떤 추세가 증가에서 감소로 변화하는 경우에, 그것을 추세의 “방향이 변화했다”라고 언급할 것입니다.

- `계절성` : 주기별로 반복되는 패턴을 의미합니다. 어떤 특정한 때나, 1주일마다 특정 요일에 나타나는 것 같은 계절성 요인이 시계열에 영향을 줄 때, 계절성 패턴이 나타난다고 말합니다. 계절성은 빈도의 형태로 나타나는데, 그 빈도는 항상 일정하며 알려져 있습니다. 

- `주기` : 고정된 빈도가 아닌 형태로 증가나 감소하는 모습을 보일 때 주기(cycle)가 나타난다고 말합니다. 보통 이러한 요동은 경제 상황 때문에 일어나고, 흔히 “경기 순환(business cycle)”과 관련 있습니다. 보통 이러한 요동의 지속기간은 적어도 2년 이상입니다.

```python
passenger = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv")
passenger['Month'] = pd.to_datetime(passenger['Month'])
passenger.set_index('Month',inplace=True)
def plot_timeseries_data(df, title ):
    plt.figure(figsize=[20,5])
    plt.plot(df)
    plt.title(title)
    plt.tight_layout()
    plt.show()
    
plot_timeseries_data(passenger, "Pan-Am항공 국제선 이용객 수 (1949~1960)")
```

- 월별로 계절성이 있다고 가정할 수 있다. (이 부분은 다양한 추론이 가능하고 이에 대한 검증을 진행하는 방향으로 분석을 진행할 수 있습니다) 

__추세__ 

- 각 연도별 합계 (아래의 그래프는 총합) 
- 추세선 긋기, 총합, 평균 등등 여러 통계치로 확인이 가능 
- 계절성을 없애는 것을 추세라고 한다. (smoothing이라고 표현하기도 한다) 
- 이동평균법도 있음 

```python
trend = passenger.reset_index()
trend['year'] = trend['Month'].dt.year
trend = trend.groupby(['year']).sum()[['Passengers']]
plot_timeseries_data(trend, "Pan-Am항공 국제선 이용객 수 (1949~1960)")
```

```python
import scipy.stats as sp
import numpy as np 

# 계졀성을 없애는 방법 
def linearRegression(df):
    y = np.array(df['Passengers'].values)
    x = np.array(df.index.values,dtype=float)
    print(y.shape,x.shape)
    slope, intercept, r_value, p_value, std_err = sp.linregress(x,y)
    xf = np.linspace(min(x),max(x),len(df.index.values))
    xf1 = xf.copy()
    xf1 = pd.to_datetime(xf1)
    yf = (slope*xf)+intercept
    print('r = ', r_value, '\n', 'p = ', p_value, '\n', 's = ', std_err)
    return xf1, yf

    
x,y_ = linearRegression(passenger)
```

```python
import seaborn as sns 

# passenger.reset_index(inplace=True)
passenger['Month'] = pd.to_datetime(passenger['Month'])

def plot_timeseries_data(df, title,trend ):
    plt.figure(figsize=[20,5])
    sns.lineplot(x= "Month", y = "Passengers",data= df)
    sns.lineplot(x = df['Month'],y=trend)
    plt.title(title)
    plt.tight_layout()
    plt.show()
plot_timeseries_data(passenger,"Pan-Am항공 국제선 이용객 수 (1949~1960)",y_)
```

### 시계열의 분해(Decomposition) 

- 이를 위해 사용하는 모델은 statsmodels 라이브러리 입니다. 
- 가법 모형 (additive model) 을 통해서 시계열 데이터를 trend + seasonality + cycle +irregular error 로 나타내어 분해해보도록 하겠습니다.
- trend와 seasonal로 설명이 잘 된다면 irregular error 는 아주 작을 것이다. 
- trend와 seasonal을 통해 예측이 가능하게 된다. 

```python
from statsmodels.tsa.seasonal import seasonal_decompose 
# 데이터 읽기 
passenger = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv")
passenger['Month'] = pd.to_datetime(passenger['Month'])
passenger.set_index('Month',inplace=True)

# ts = passenger.index
# def plot_decomposition(df):
#     result = seasonal_decompose(df['Passengers'], model = "additive")
#     # seasonal 12개월 
#     df['Trend'] = result.trend
#     df['Seasonal'] = result.seasonal
#     df['Resid'] = result.resid
    
#     fig,ax = plt.subplots(nrows=4,ncols=1,figsize= [20,10],sharex=True)
#     df[['Passengers']].plot(ax=ax[0])
#     df[['Trend']].plot(ax=ax[1])
#     df[['Seasonal']].plot(ax=ax[2])
#     df[['Resid']].plot(ax=ax[3])
#     plt.tight_layout()
#     plt.show()
result = seasonal_decompose(passenger['Passengers'], model = "additive")

plt.rcParams['figure.figsize'] = [12, 8]
result.plot()
plt.show()


# plot_decomposition(passenger)
```
